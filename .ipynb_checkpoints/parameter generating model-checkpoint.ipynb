```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('final_data_set.csv')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
import warnings
warnings.filterwarnings('ignore')
for i in range(len(data.humidity)):
     data['date'][i]=data['date'][i].replace('-','')
```

```python
x=data.iloc[:,0]
y=data.iloc[:,1:4]
```

```python
li_model=LinearRegression(fit_intercept=True, copy_X=True, n_jobs=None, positive=False)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=0)
```

```python
li_model.fit(np.array(x_train).reshape(-1,1),y_train)
```

```python
plt.plot(li_model.predict(np.array(x_test).reshape(-1,1)),c='b')
```
